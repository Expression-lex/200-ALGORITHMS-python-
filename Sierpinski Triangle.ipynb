```python
from random import choice
import matplotlib.pyplot as plt

%matplotlib inline
```

# Sierpinski Triangle

First Transformation 
\begin{equation}
x_{n+1} = 0.5x_n 
\end{equation}
\begin{equation}
y_{n+1} = 0.5y_n
\end{equation}
Second Transformation
\begin{equation}
x_{n+1} = 0.5x_n + 0.5
\end{equation}
\begin{equation}
y_{n+1} = 0.5y_n + 0.5
\end{equation}
Third Transformation
\begin{equation}
x_{n+1} = 0.5x_n + 1
\end{equation}
\begin{equation}
y_{n+1} = 0.5y_n
\end{equation}


```python
def trans_1(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x
    y1 = 0.5 * y

    return x1, y1

def trans_2(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 0.5
    y1 = 0.5 * y + 0.5

    return x1,y1

def trans_3(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 1
    y1 = 0.5 * y

    return x1, y1

transformations = [trans_1, trans_2, trans_3]

a1 = [0]
b1 = [0]
a,b = 0,0

for i in range(100000):
    trans = choice(transformations)
    a,b = trans((a,b))
    a1. append(a)
    b1. append(b)
```

```python
plt.rc('figure', figsize =(16,16))
```

```python
plt.plot(a1,b1,'o')
plt.savefig('my_figure.png')
```
